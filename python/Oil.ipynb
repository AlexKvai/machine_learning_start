```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_excel('gdprussia.xlsx')
```

```python
df.head()
```

```python
%matplotlib inline
plt.scatter(df.oilprice, df.gdp)
plt.xlabel('oilprice (US$)')
plt.ylabel('gdp (bln US$)')
```

```python
reg = linear_model.LinearRegression()

reg.fit(df[['oilprice']], df.gdp)
```

```python
reg.predict(df[['oilprice']])
```

```python
%matplotlib inline
plt.scatter(df.oilprice, df.gdp)
plt.xlabel('oilprice (US$)')
plt.ylabel('gdp (bln US$)')
plt.plot(df.oilprice, reg.predict(df[['oilprice']]), color='red')
```

```python
reg.predict([[30]])
```

```python
reg = linear_model.LinearRegression()

reg.fit(df[['year','oilprice']], df.gdp)
```

```python
reg.predict(df[['year','oilprice']])
```
